```python
# Импортируйте все необходимые библиотеки и модули которые вам необходимы
import numpy as np
import pandas as pd
```

```python
# Загрузите датасет и проведите его очистку и форматирование если это необходимо
file_path = r'baseball_games.csv'
#предусмотрительно заглянув в файл, обнаруживаем что есть колонка индексов 'Unnamed: 0' и можно использовать ее 
#для индексации, чтобы не плодить новых колонок
baseball_df = pd.read_csv(file_path, index_col='Unnamed: 0') 
baseball_df.head()
```

```python
baseball_df.describe()
```

```python
baseball_df.info()
```

По полю attendance количество отличается от количества записей в dataframe, значит есть null значения, заменим их на 0, предположив, что зрителей на этих матчах не было. Можно было бы дропнуть, но возможно есть полезная инфа по другим колонкам. 

```python
baseball_df.fillna(0, inplace=True)
baseball_df.info()
```

# 7. Какая игра была последней в сезоне?

по одному только полю date мы нужный ответ не получаем, поэтому я добавила новую колонку где соединила дату и время начала в одно значение

```python
#заменила p.m.|a.m. на аббревиаттуры, которые понимает to_datetime, и отбросила слово Local, т.к. оно в каждой строке.
baseball_regular.start_time.replace(regex=True, inplace=True, to_replace=['p.m.', 'a.m.', 'Local'],value=['PM', 'AM', ''])
baseball_regular['start_date_time'] = pd.to_datetime(baseball_regular['start_time']+baseball_regular['date'])
```

```python
baseball_regular[baseball_regular['start_date_time']==baseball_regular['start_date_time'].max()]
```

# 14. Какая команда выиграла наибольшое количество матчей в сезоне? (Будьте внимательны с типом матча в сезоне - season).

```python
# baseball_regular.iloc[
home_teams = baseball_regular[['home_team','home_team_win', 'home_team_loss']].pivot_table(values=['home_team_win', 'home_team_loss'], 
                                                                              index=['home_team' ],
                                                                              aggfunc='sum')
home_teams.rename(columns={'home_team':'team', 'home_team_win':'win', 'home_team_loss':'loss'}, inplace=True)
away_teams = baseball_regular[['away_team','home_team_win', 'home_team_loss']].pivot_table(values=['home_team_loss', 'home_team_win'], 
                                                                              index=['away_team' ],
                                                                              aggfunc='sum')
away_teams.rename(columns={'away_team':'team', 'home_team_win':'loss', 'home_team_loss':'win'}, inplace=True)
teams_results = home_teams+away_teams
teams_results.iloc[[teams_results['win'].argmax()]]
```

# 15. Какая команда выиграла наибольшее количество домашних матчей в сезоне?

```python
temp_df = pd.DataFrame(baseball_regular[['home_team', 'home_team_win']].groupby(by=['home_team']).sum())
temp_df.iloc[temp_df['home_team_win'].argmax()]
```

# 16. Какая команда выиграла наибольшее количество гостевых матчей в сезоне?

```python
temp_df = pd.DataFrame(baseball_regular[['away_team', 'home_team_loss']].groupby(by=['away_team']).sum())
temp_df.iloc[temp_df['home_team_loss'].argmax()]
```

# 17. Какая команда проиграла наибольшее количество матчей в сезоне?

```python
teams_results.iloc[[teams_results['loss'].argmax()]]
```

# 18. Зависит ли выигрыш от количества посетителей матча? (https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.cov.html)

```python
pd.set_option('display.float_format', lambda x: '%.3f' % x)
baseball_regular.cov()
```

учитывая то, что ковариация между attendance и home_team_win положительная, следовательно положительная связь присутствует и статистически выйгрыш матча зависит от количества посетителей.

# 19. Правда ли что большинство проигрышных домашних матчей приходятся на Субботу и Воскресенье?

```python
baseball_regular[['day_of_week', 'home_team_loss']].groupby(by=['day_of_week']).sum().sort_values(by='home_team_loss').iloc[-2:]
```
